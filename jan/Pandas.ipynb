### Indexing a Dataframe


```python
# loading a dataframe
df3 = pd.read_csv("./read_csv/cancer.csv")

# display the first 5 rows
df3.head()
```

```python
df3.columns
```

```python
# displaying multliple columns

df3[["Cell_Shape_Uniformity", "Marginal_Adhesion"]].head()
```

```python
# more columns
df3[["Cell_Shape_Uniformity", "Marginal_Adhesion", "Bare_Nuclei"]].tail()
```

```python
# displaying rows : using iloc function in dataframe
df3.iloc[:5]
```

```python
# selecting columns using the iloc function
# the first : is the row i.e all rows
# the second : is for all columns, :2 means from 0 to 1
df3.iloc[:,:2]
```

```python
# subsetting the dataframe
# selecting a row based on some condition
# eg selecting the row where the cancer type is equal to 2

df3[df3["Cancer_Type"]==2]
```
